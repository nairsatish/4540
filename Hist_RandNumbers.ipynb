```python
#ChatGPT - generate random numbers between -100 to 100 with mean of 5 and SD of 2, and make a histogram plot with 50 bins
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters for the normal distribution
mean = 5
std_dev = 2
num_samples = 1000000

# Generate random numbers from a normal distribution
random_numbers = np.random.normal(mean, std_dev, num_samples)

# Ensure that the generated numbers fall within the desired range [0, 10]
random_numbers = np.clip(random_numbers, -100, 100)

# Plot the histogram
plt.hist(random_numbers, bins=50, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Histogram of Random Numbers with Normal Distribution (mean=5, std=2)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
