```python
import numpy as np
from scipy.stats import norm
import scipy.signal as signal
import matplotlib.pyplot as plt
```

Generate random vars with normal distribution, mean=0, sigma=1, and of size =1000

```python
x = norm.rvs(size=1000) # generate random samples
```

```python
# Plot histogram and probability density function

plt.figure()
_ , bins, _ = plt.hist(x,bins=30,density=True)
plt.plot(bins,norm.pdf(bins),'r')
plt.xlabel('x')
plt.ylabel('probability density')
plt.show()
```

```python
fs = 1000 # sampling frequency of time series (Hz)
t = np.arange(x.size)/fs*1000 # time points (ms)
#print(t)
```

```python
# White noise

plt.figure()
plt.plot(t, x)
plt.xlabel('t (ms)')
plt.ylabel('x(t)')
plt.show()
```

```python
# Calculate power spectral density using welch's method
f, pxx = signal.welch(x, fs=fs)

plt.figure()
plt.plot(f,pxx)
plt.xlabel('frequencies (Hz)')
plt.ylabel('PSD')
plt.show()
```

```python
b = np.cumsum(x) # Brownian noise
plt.figure()
plt.plot(t,b)
plt.xlabel('t(ms)')
plt.ylabel('Brownian noise')
plt.show()
```
